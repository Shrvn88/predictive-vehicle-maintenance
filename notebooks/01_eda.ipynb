```python
import matplotlib.pyplot as plt

sample_units = train_df["unit"].unique()[:5]

plt.figure(figsize=(10,5))

for u in sample_units:
    sub = train_df[train_df["unit"] == u]
    plt.plot(sub["cycle"], sub["RUL"])

plt.xlabel("Cycle")
plt.ylabel("RUL")
plt.title("RUL Degradation Curves")
plt.show()
```

```python
for i in range(1, 22):
    sensor = f"sensor_{i}"

    plt.figure(figsize=(10,5))

    for u in sample_units:
        sub = train_df[train_df["unit"] == u]
        plt.plot(sub["cycle"], sub[sensor])

    plt.title(sensor)
    plt.show()
```
